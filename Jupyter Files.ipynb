```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape 
#use for showing data lodaed or not
```

```python
df.head()
#usigng showing first 5 rows of loaded data
```

```python
df.info()
#Showing the details of data
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
#Detele unnecessary columns
```

```python
pd.isnull(df).sum()
#give total number of null values count
```

```python
df.dropna(inplace=True)
#Delete the Null valurs
```

```python
df['Amount'] = df['Amount'].astype('int')
#Change data type
```

```python
df.columns
```

```python
df.describe()
#describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
```

```python
# Exploratory Data Analysis
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)

# plotting a bar chart for Gender and it's count
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)

# plotting a bar chart for gender vs total amount
```

```python
df['Gender'].value_counts()
```

```python
df['Age Group'].value_counts()
```

```python
df['State'].value_counts()
```

```python
sns.pairplot(df, hue='Gender')  
plt.title("Pairplot of Customer Data")  
plt.show()

#This visual helps identify relationships between different numerical variables.
#Look for clustering based on gender,
#Which indicates possible differences in spending behavior.
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])  
correlation_matrix = numeric_df.corr() 
plt.figure(figsize=(10, 6))  
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)  
plt.title("Correlation Heatmap")  
plt.show() 
```

```python
avg_spending_by_state = df.groupby('State')['Amount'].mean().sort_values(ascending=False)
print(avg_spending_by_state)
#Groupby Analysis:Use groupby in pandas to analyze average spending by State or Occupation.
```

```python
df['Amount'].hist(bins=5)  
plt.title("Distribution of Purchase Amount")  
plt.xlabel("Amount")  
plt.ylabel("Frequency")  
plt.show()  

#The histogram of the Amount shows the frequency of various purchase amounts. 
#You can observe if most customers are clustering within a certain spending range.
```

```python
sns.boxplot(x='Gender', y='Amount', data=df)  
plt.title("Boxplot of Amount by Gender")  
plt.show()  
#The boxplot indicates the distribution of the purchase amount by gender. 
#If one gender has higher median amounts or wider interquartile ranges.
#It might suggest spending behavior differences.
```

```python
sns.scatterplot(data=df, x='User_ID', y='Amount', hue='Gender')  
plt.title("Scatterplot of User_ID vs Amount")  
plt.show()  
#This plot can reveal if younger or older tend to spend more. Clustering may indicate trends in age-based spending.
```
